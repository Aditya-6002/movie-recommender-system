```python
def rating(x, C=C, m=m):
    v = x["vote_count"]
    r = x["vote_average"]
    return (v/(v+m) * r) + (m/(v+m) * C)
```

```python
custom_stopwords = list(text.ENGLISH_STOP_WORDS.union({'boom'})) # This has all the enlish words like and , or ,  they because these words make the process slow and have no big value 
 # see this is custom word if you want you can use this method too remove a custom word too

tfidf = TfidfVectorizer(stop_words=custom_stopwords)

movie["overview"] = movie["overview"].fillna(" ")

tfidf_matrix = tfidf.fit_transform(movie["overview"])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
import pandas as pd

indices = pd.Series(movie.index, index=movie['title']).drop_duplicates()

def get_recommendations(title):
    idx = indices[title]

    cosine_sim = linear_kernel(tfidf_matrix[idx:idx+1], tfidf_matrix)

    sim_scores = list(enumerate(cosine_sim[0]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:11]

    movie_indices = [i[0] for i in sim_scores]

    return movie.iloc[movie_indices][['title', 'release_date', 'vote_average']]
```

```python
get_recommendations("The Dark Knight Rises")
```

```python
# literal_eval = turn string into list so that we can extract the given data easily

from ast import literal_eval
features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    movie[feature] = movie[feature].apply(literal_eval)
```

```python
import numpy as np
def director(x):
    for i in x:
        if i["job"]=="Director":
            return i["name"]
    return np.nan
          
    
```

```python
def get_list(x):
    if isinstance(x,list):
        return[i["name"] for i in x[:3]]
    else:
        return[]
```

```python
movie["director"] = movie["crew"].apply(director)
features = ['cast', 'keywords', 'genres']
for feature in features:
    movie[feature] = movie[feature].apply(get_list)
```

```python

movie[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    movie[feature] = movie[feature].apply(clean_data)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
```

```python

movie['soup'] = movie.apply(create_soup, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(movie['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_recommendations(title):
    idx = indices[title]

    
    sim_scores = cosine_similarity(count_matrix[idx], count_matrix).flatten()

    sim_scores = list(enumerate(sim_scores))

   
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

   
    sim_scores = sim_scores[1:11]

    movie_indices = [i[0] for i in sim_scores]

    return movie['title'].iloc[movie_indices]
```

```python
movie = movie.reset_index(drop=True)
indices = pd.Series(movie.index, index=movie['title'])
```
